Import Library

```python
from nltk.tokenize import RegexpTokenizer
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

Dataset

```python
dataset = {}
dataset["D1"] = "lari lari pagi itu bikin sehat, badan sehat itu penting.\n"
dataset["D2"] = "pagi pagi kok sudah lari pagi\n"
dataset["D3"] = "korupsi akan sulit hilang, karena korupsi sudah jadi budaya\n"
dataset["D4"] = "agar sehat harus banyak lari lari pagi\n"
dataset["D5"] = "hobi kita korupsi ,budaya ya budaya\n"
```

Case Folding

```python
for k in dataset.keys():
  hasil_case_folding = dataset[k].lower()
  dataset[k] = hasil_case_folding
  print(dataset)
```

Tokenizing


```python
tokenizer = RegexpTokenizer(r'\w+')
for k in dataset.keys():
  tokens = tokenizer.tokenize(dataset[k])
  dataset[k] = tokens
  print(dataset)
```

Stopword by Sastrawi

```python
factory = StopWordRemoverFactory()
stopword_list=factory.get_stop_words()
for k in dataset.keys():
  tokens=[]
  for t in dataset[k]:
    if t not in stopword_list:
      tokens.append(t)
  dataset[k]=tokens
print(dataset)
```

Mencari Kata Uniq

```python
listku=list(dict.fromkeys(tokens))
print(listku)
```

Dataset

```python
dataset = {}
dataset["D1"] = "lari lari pagi itu bikin sehat, badan sehat itu penting.\n"
dataset["D2"] = "pagi pagi kok sudah lari pagi\n"
dataset["D3"] = "korupsi akan sulit hilang, karena korupsi sudah jadi budaya\n"
dataset["D4"] = "agar sehat harus banyak lari lari pagi\n"
dataset["D5"] = "hobi kita korupsi ,budaya ya budaya\n"
dataset["q"] = "lari pagi budaya"
```

Frekuensi Kemunculan Kata

```python
tf = CountVectorizer()
term_doc_matrix = tf.fit_transform(dataset.values())
pd.DataFrame(term_doc_matrix.toarray(), index=dataset.keys(), columns=tf.get_feature_names_out())
```

Perhitungan Eucliden Distance

```python
ed = euclidean_distances(term_doc_matrix, term_doc_matrix)
df_ed=pd.DataFrame(ed, index=dataset.keys(), columns=dataset.keys())
df_ed
```

Pengurutan Ranking Euclidean Distance

```python
rank_ed={}
for k in dataset.keys():
  if k !="q":
    rank_ed[k]=df_ed.at[k,"q"]
top_rank_ed=dict(sorted(rank_ed.items(), key=operator.itemgetter(1)))
pd.DataFrame(top_rank_ed.values(),index=top_rank_ed.keys(), columns=["Euclidean Distance"])
```

Pembobotan TF.IDF

```python
tfidf=TfidfVectorizer()
inverted_index=tfidf.fit_transform(dataset.values())
pd.DataFrame(inverted_index.toarray(), index=dataset.keys(), columns=tfidf.get_feature_names_out())
```

Perhitungan Cosine Similarity

```python
cs = cosine_similarity(inverted_index,inverted_index)
df_cs = pd.DataFrame(cs, index=dataset.keys(), columns=dataset.keys())
df_cs
```

Pengurutan Ranking Cosine Similarity

```python
rank_cs = {}
for k in dataset.keys():
  if k !="q":
    rank_cs[k]=df_cs.at[k,"q"]
top_rank_cs = dict(sorted(rank_cs.items(), key=operator.itemgetter(1), reverse=True))
pd.DataFrame(top_rank_cs.values(), index=top_rank_cs.keys(), columns=["Cosine Similarity"])
```
